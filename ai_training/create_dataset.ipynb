```python
import numpy as np
import matplotlib.pyplot as plt
from dataset import TrainingData
```

```python
training_data = TrainingData(
    filter="F814W", pixel_size=0.04, num_pixel=128, verbose=False, source_type="galaxy"
)
```

```python
num_systems = 10
dataset, masks = training_data.create_dataset(
    num_systems, max_satellite_num=3, no_lens_light_fracion=0.2
)
```

```python
fig, axes = plt.subplots(num_systems, 2, figsize=(10, 4 * num_systems))

for i in range(num_systems):
    axes[i][0].matshow(np.log10(dataset[i]), cmap="cubehelix", vmin=-3)
    im = axes[i][1].matshow(masks[i], vmax=4)

    axes[i][0].axis("off")
    axes[i][1].axis("off")

axes[0][0].set_title("Image")
axes[0][1].set_title("Mask")
```
